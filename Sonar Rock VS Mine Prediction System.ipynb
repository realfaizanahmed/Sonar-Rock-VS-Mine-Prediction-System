```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('sonar.csv', header=None)
```

```python
sonar_data.head()
```

```python
# The last column is the target variable, and the rest are features
X = sonar_data.iloc[:, :-1]
Y = sonar_data.iloc[:, -1]
```

```python
# Convert the target variable to binary (Rock: 0, Mine: 1)
Y = Y.map({'R': 0, 'M': 1})
```

```python
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
#Train the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
#Predict on the test data
Y_pred = model.predict(X_test)
```

```python
#Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Function to take user input for prediction
def get_user_input():
    user_input = input(f"Enter {X.shape[1]} feature values separated by spaces or commas: ")
    input_data = np.array([float(i) for i in user_input.replace(',', ' ').split()])
    return input_data.reshape(1, -1)

# Get user input
user_input = get_user_input()

# Scale the input data (using the same scaler used for training data)
user_input_scaled = scaler.transform(user_input)

# Make a prediction
prediction = model.predict(user_input_scaled)

# Output the prediction result
if prediction[0] == 0:
    print('The object is predicted to be a Rock.')
else:
    print('The object is predicted to be a Mine.')
```
